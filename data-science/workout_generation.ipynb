```python
import pandas as pd
import random
```

```python
df = pd.read_csv('../data-science/dataset/dataset_exercices_crosstraining.csv')
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df = df.applymap(lambda x: x.strip() if type(x) == str else x)
df['Muscles Worked'] = df['Muscles Worked'].str.replace('and|\\sand\\s',' ',regex=True)

# Use Series.replace with regex=True
df['Muscles Worked'] = df['Muscles Worked'].replace({r'quads(\s|$)': r'quadriceps\1',r'full-body(\s|$)': r'full_body\1',r'hamstrings(\s|$)': r'hamstring\1',
                r'shoulders(\s|$)': r'shoulder\1',r'glutes(\s|$)': r'glute\1',r'hams(\s|$)': r'hamstring\1'}, regex=True)
```

```python
# Import defaultdict
from collections import defaultdict

# Create an empty defaultdict with list as default value
muscle_dict = defaultdict(list)

# Iterate over the rows of df
for index, row in df.iterrows():
  # Get the exercise name and the list of muscle groups
  exercise = row['Name']
  muscles = row['Muscles Worked']
  # For each muscle group in the list, append the exercise name to its value in muscle_dict
  for muscle in muscles.split(" "):
    muscle_dict[muscle].append(exercise)


random.choices(muscle_dict['chest'])
list(muscle_dict.keys())
```

```python
import numpy as np

exercises = muscle_dict
def generate_workouts(target):
  
  # Create an empty list to store the workouts
  workouts = []

  # Iterate over the target dictionary
  for muscle_group in target:

    # Get the number of sets and reps for the current muscle group
   # sets_reps = target[muscle_group]

    # Get the possible exercises for the current muscle group from the exercises dictionary
    possible_exercises = exercises[muscle_group]


    # Choose one exercise randomly based on their weights or scores using random.choices function
    #chosen_exercise = random.choices(list(possible_exercises), weights=list(possible_exercises), k=1)[0]
    chosen_exercise = random.choices(possible_exercises)[0]

    # Append a sublist with the chosen exercise name and number of sets and reps to the workouts list
    #workouts.append([chosen_exercise] + [sets_reps])
    workouts.append(chosen_exercise)

  
   # Shuffle or randomize the order of the workouts list using random.shuffle function 
  random.shuffle(workouts)

  # Return the workouts list 
  return (workouts)
     

def wod_composition(num_exercices):
    type_wod=['emom', 'amrap', 'fortime', 'tabata']
    workout_compo = {}
    duration = 0

    wod_type = random.choice(type_wod)
    if wod_type =='emom':
        duration = random.choice(np.arange(6, 16, num_exercices))
    elif wod_type=='amrap':
        duration = random.choice(np.arange(6, 16, num_exercices))
    elif wod_type=='fortime':
        print('no duration limit')
    elif wod_type=='tabata':
        duration = random.choice(np.arange(6, 16, num_exercices))

    workout_compo[wod_type] = ( generate_workouts(list(muscle_dict.keys())), duration)
    return workout_compo



def generate_wod(duration):
        #duration =1, 2 ou 3 number of 'emom', 'amrap', 'fortime' or 'tabata'
        workout=[]
        for i in range(duration):
            num_exercises = random.randint(1,duration)
            workout.append(wod_composition(num_exercises))
        return(workout)

print((generate_wod( 4)))       
```
